```python
#test data is hidden
from sklearn.model_selection import train_test_split
import pandas as pd

data = {
    "rooms": [2,3,4,5,3,4,2,5],
    "income": [40000,50000,60000,70000,52000,65000,42000,75000],
    "price": [200000,300000,400000,500000,320000,450000,220000,520000]
}

df = pd.DataFrame(data)

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)

train_set, test_set
```

```python
#graph to spot pattern
import matplotlib.pyplot as plt

plt.scatter(df["income"], df["price"])
plt.xlabel("Income")
plt.ylabel("House Price")
plt.show()
df.corr()
```
